Projeto Python IA: Inteligência Artificial e Previsões 

Case: Score de Crédito dos Clientes

Contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base na analise,
criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, OK, Bom.      


```python
# Passo a Passo 
# Passo 0 - Entender o desafio da empresa 
# Passo 1 - Importar a base de dados 
# Passo 2 - Prepara a base de dados para a inteligência antificial 
# Passo 3 - Criar um modelo de IA -> Score de crédito: Bom, Médio, Ruim
# Passo 4 - Escolher o melhor modelo
# Passo 5 - Usar a nossa IA para fazer novas previsões
# Bibliotecas: 
#!pip install pandas numpy scikit-learn 
```

```python
import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
print(tabela.info())


```

```python
#Coluna profissao, comportamento_pagamento e score_credito são colunas de texto. 
#Como converter coluna de dados do tipo texto em número: 
#Exemplo: Inserir números na frente das profissões: engenheiro - 1 / cientista - 2 /  analista - 3 / professor - 4 / estudante - 5 / enfermeiro - 6 / medico - 7 / advogado - 8 / tecnico - 9
#Processo chamado de ( LabelEncoder ) - Ferramenta vem do sklearn 
#Utilizando apenas o LabelEncoder que está no sklearn 
from sklearn.preprocessing import LabelEncoder 

codificador = LabelEncoder()

# codificador -> aplica na coluna "profissao"
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])

# codificador -> aplica na coluna "comportamento_pagamento"
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

print(tabela.info())


```

```python
#Processo/Aprendizado de máquina 

# Divisão de Y (Coluna que você quer prever)
# Divisão de X (Coluna que você quer para fazer a previsão)
    # Não vamos usar a coluna id_clientes porque ela é um número aleatório

y = tabela["score_credito"]
x = tabela.drop(columns=["score_credito", "id_clientes"])

# utilizando o sklearn a biblioteca de treino e teste para x e y. Para realizar na I.A de modo que ela aprenda e não apenas decore. 
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
#Criar a Inteligência Artificial 
# Árvore de Decisão -> RandomForest 
    # Explicação: Ele faz perguntas para a base de dados. 
    
# KNN -> Vizinhos Próximos -> Kneighbors 
    # Pega todos os clientes e joga em um gráfico, realizando comparações de outros cliente que estão no mesmo nível, ou em níveis diferentes e os classificando. 

# Importar a inteligencia antificial 
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar a inteligencia antificial 
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Treinar a inteligencia antificial
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
 
```

```python
# Testar os modelos / comparar os modelos 
#descobrindo o melhor modelo para I.A
# Acurácia 
from sklearn.metrics import accuracy_score

previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))

# Neste caso a melhor I.A a ser utilizando é (KNN -> Vizinhos Próximos -> Kneighbors )

```

```python
# Melhor modelo: modelo_arvoredecisao 
# Fazer novas previsões 
# Importar novos clientes 

tabela_novos_clientes = pd.read_csv("novos_clientes.csv")
display(tabela_novos_clientes)

# Remover colunas desnecessárias
tabela_novos_clientes = tabela_novos_clientes.drop(['id_clientes', 'score_credito'], axis=1)

# Codificar os novos clientes 
# codificador -> aplica na coluna "profissao"
tabela_novos_clientes["profissao"] = codificador.fit_transform(tabela_novos_clientes["profissao"])

# codificador -> aplica na coluna "comportamento_pagamento"
tabela_novos_clientes["comportamento_pagamento"] = codificador.fit_transform(tabela_novos_clientes["comportamento_pagamento"])

# Fazer as previsões
previsoes = modelo_arvoredecisao.predict(tabela_novos_clientes)
print(previsoes)
```
